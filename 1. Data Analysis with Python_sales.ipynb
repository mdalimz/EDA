# 4.1 Numerical visualization

```python
sales['Unit_Price'].plot(kind='density',figsize=(14,6))
```

```python
sales['Profit'].plot(kind='hist', figsize=(14,8))
```

```python
ax = sales['Unit_Cost'].plot(kind='density', figsize=(14,6)) # kde
ax.axvline(sales['Unit_Cost'].mean(), color='purple')
ax.axvline(sales['Unit_Cost'].median(), color='red')
ax.axvline(sales['Unit_Cost'].std(), color='green')
ax.set_ylabel('Number of Sales')
ax.set_xlabel('dollars')
plt.legend(['unit_cost', 'mean', 'median', 'std'])
```
